# Outliers Detection & removal using Percentile

```python
e_com_data.head()
```

```python
for i in e_com_data:
    min_threshold,max_threshold=e_com_data[i].quantile([0.05,0.95])   #taking threshold value 95%
    
    #upper_limit & lower_limit
    print(min_threshold,max_threshold) 
    
    #data which is above the upper limit(capping)
    print(e_com_data[e_com_data[i]>max_threshold].shape)
    
    #data which is below the lower limit(flooring)
    print(e_com_data[e_com_data[i]<min_threshold].shape)
    
    
    #new data column : replacing outliers with lower and uper limit
    e_com_data[i]=np.where(e_com_data[i]>=max_threshold,
                                         max_threshold,
                                         np.where(e_com_data[i]<=min_threshold,
                                                 min_threshold,
                                                 e_com_data[i])
                          )
```

# 1) Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error,r2_score,accuracy_score

linear_model_1 = LinearRegression()
linear_model_1.fit(X_train,y_train)
```

```python
print('intercepr : ',linear_model_1.intercept_,'\ncoefficients :',linear_model_1.coef_)
```

```python
prediction=linear_model_1.predict(X_test)
print(linear_model_1.score(X_train, y_train).round(4))
print(linear_model_1.score(X_test, y_test).round(4))
```

**Training Data Accuracy**

```python
# Measuring accuracy on training data
def model_accuracy_train(model_name):
    print(model_name,
          '\n******************************************************************************************',
          '\nmean absolute error             : ',mean_absolute_error(y_train,model_name.predict(X_train)),
          '\nmean_absolute_percentage_error  : ',mean_absolute_percentage_error(y_train,model_name.predict(X_train)),
          '\nr2_score                        : ',r2_score(y_train,model_name.predict(X_train)),
          '\nmean squared error              : ',mean_squared_error(y_train, model_name.predict(X_train)),
          '\nRoot Mean Square Error          : ',np.sqrt(mean_squared_error(y_train, model_name.predict(X_train))),
          '\n******************************************************************************************')
```

```python
model_accuracy_train(linear_model_1)
```

**Test Data Accuracy**

```python
# Measuring accuracy on testing data
def model_accuracy(model_name):
    print(model_name,
          '\n******************************************************************************************',
          '\nmean absolute error             : ',mean_absolute_error(y_test,model_name.predict(X_test)),
          '\nmean_absolute_percentage_error  : ',mean_absolute_percentage_error(y_test,model_name.predict(X_test)),
          '\nr2_score                        : ',r2_score(y_test,model_name.predict(X_test)),
          '\nmean squared error              : ',mean_squared_error(y_test, model_name.predict(X_test)),
          '\nRoot Mean Square Error          : ',np.sqrt(mean_squared_error(y_test, model_name.predict(X_test))),
          '\n******************************************************************************************')
```

```python
model_accuracy(linear_model_1)
```
