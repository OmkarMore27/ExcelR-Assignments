# Text Preprocessing


```python
import re
import string
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer() 
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
def preprocess(sentence):
    sentence=str(sentence)
    sentence = sentence.lower()
    sentence=sentence.replace('{html}',"") 
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url=re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num)  
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
    #stem_words=[stemmer.stem(w) for w in filtered_words]
    #lemma_words=[lemmatizer.lemmatize(w) for w in stem_words]
    return " ".join(filtered_words)
```

```python
text_data['cleanText']=text_data['content'].map(lambda s :preprocess(s)) 
```

```python
text_data[['cleanText']]
```

```python
def lemmatise(sentence):
  tokenizer = RegexpTokenizer(r'\w+')
  tokens = tokenizer.tokenize(sentence) 
  #stem_words=[stemmer.stem(w) for w in tokens]
  lemma_words=[lemmatizer.lemmatize(w) for w in tokens]
  return " ".join(lemma_words)
```

```python
tokenizer = RegexpTokenizer(r'\w+')
text_data['lemmatized_content']=text_data['cleanText'].map(lambda x : tokenizer.tokenize(lemmatise(x)))
text_data['lemmatized_content']
```

```python
#Installing the gensim library required for word2Vec 

import gensim
#Loading the word vectors from Google trained word2Vec model
GoogleModel = gensim.models.KeyedVectors.load_word2vec_format('/content/drive/MyDrive/SCIENTIST/GOOGLE_NEWSDATASET/GoogleNews-vectors-negative300.bin', binary=True,)
```

```python
WordsVocab=VectorizedData.columns
WordsVocab
```

```python
len(WordsVocab)
```

```python
## WORD2VEC CONVERSION
def FunctionText2Vec(TextData):
    # transform the text to numeric data
    X = vectorizer.transform(TextData)
    CountVecData=pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
    
    # Make empty dataframe to hold sentences
    W2Vec_Data=pd.DataFrame()
    
    # Looping through each row for the data
    for i in range(CountVecData.shape[0]):

        # initiating a sentence with all zeros
        Sentence = np.zeros(300)

        # Looping through each word in the sentence and whether its present in 
        # the Word2Vec model then storing its vector
        for word in WordsVocab[CountVecData.iloc[i,:]>=1]:
            #print(word)
          if word in GoogleModel.wv.vocab:    
                Sentence=Sentence+GoogleModel[word]
        # Appending the sentence to the dataframe
        W2Vec_Data=W2Vec_Data.append(pd.DataFrame([Sentence]))
    return(W2Vec_Data)
```

```python

# Calling the function to convert all the text data to Word2Vec Vectors.I Passed all my Descriptive Feature for Vectorization
W2Vec_Data=FunctionText2Vec(text_data['cleanText'])
 
# Checking the new representation for sentences
W2Vec_Data.shape 
```
